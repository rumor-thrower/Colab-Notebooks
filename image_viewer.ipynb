```python
import matplotlib.pyplot as plt
```

```python
def view_number_gray_image(img):
  plt.figure(figsize=(10,10))
  plt.axis('off')
  plt.imshow(img, cmap='gray', aspect='auto')
  for i, y in enumerate(img):
    for j, x in enumerate(y):
      plt.annotate(str(img[i,j]),(j,i), ha='center', va='center', color='white' if img[i,j]<=128 else 'black')
```

```python
def view_number_color_image(img):
  plt.figure(figsize=(13,13))
  plt.axis('off')
  plt.imshow(img, cmap='gray', aspect='auto')
  for i, y in enumerate(img):
    for j, x in enumerate(y):
      plt.annotate(f"({img[i,j,0]},{img[i,j,1]},{img[i,j,2]})",(j,i), ha='center', va='center', color='black', fontsize=4, rotation=45)
```

```python
import tensorflow as tf
```

```python
(X_train, y_train), (X_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
img = X_train[0]
```

```python
view_number_gray_image(img)
```
